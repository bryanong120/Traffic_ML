```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import LabelEncoder
```

```python
# read data from csv file
accident_data = pd.read_csv('US_Accidents_Dec21_updated.csv')
```

```python
features = accident_data.iloc[:1000, [2,25,27]].values
result = accident_data.iloc[:1000, 1].values

for x in features:
    
    time = x[0].split()[1]
    time = time.replace(":","")
#     print(time)
    x[0] = int(time)
    if(x[0] is None):
        print(x)
    if(np.isnan(x[2])):
        x[2] = 0
        
    if(np.isnan(x[1])):
#         print("Is NAN")
#         print(x)
#         features = np.delete(features,i,0)
        x[1] = 0
    print(x)
```

```python
features_Train, features_Test, result_Train, result_Test = train_test_split(features, result, test_size = 0.25, random_state = 0)
```

```python
# fitting the data (train a model)
classifier = SVC(kernel='rbf',random_state=1,C=1,gamma='auto')
classifier.fit(features_Train,result_Train)
```

```python
# perform prediction on x_test data
result_pred = classifier.predict(features_Test)
```

```python
cm = confusion_matrix(result_Test,result_pred)
print(cm)
accuracy = float(cm.diagonal().sum())/len(result_Test)
print('model accuracy is:',accuracy*100,'%')
```
